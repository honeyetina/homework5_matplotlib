```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


# Combine the data into a single dataset


# Display the data table for preview
```

```python
data = pd.merge(mouse_metadata,study_results, how='inner', on='Mouse ID', )
data
```

```python
# Checking the number of mice.
number_of_mice = data['Mouse ID'].nunique()
print(number_of_mice)

```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupicate_mice_ID = data[data[['Mouse ID','Timepoint']].duplicated()]['Mouse ID'].drop_duplicates()
dupicate_mice_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

dup_data_ID = data[data['Mouse ID']=='g989']
dup_data_ID
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = data[data['Mouse ID']!='g989']
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.
number_of_clean_mice = clean_data['Mouse ID'].nunique()
number_of_clean_mice

```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

```

```python
data1_mean = clean_data.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].mean()
data1_mean
data1_median = clean_data.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].median()
data1_var = clean_data.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].var()
data1_std = clean_data.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].std()
data1_sem = clean_data.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].sem()
data1_v2 = pd.concat([data1_mean, data1_median,data1_var,data1_std,data1_sem], axis=1)
data1_v2
data1_v2.columns = ['mean','median','var','std','sem']
data1_v2
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
```

```python
data1 = clean_data.groupby(['Drug Regimen']).agg({"Tumor Volume (mm3)":['mean','median','var','std','sem']})
data1.columns = ['mean','median','var','std','sem']
data1
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
data9 = clean_data[clean_data['Drug Regimen']=='Capomulin']
data9
data10 = data9\
    .groupby(['Mouse ID', 'Weight (g)'])\
    .agg({'Tumor Volume (mm3)': 'mean'})\
    .rename(columns ={'Tumor Volume (mm3)':'Avg. Tumor Volume (mm3)'} )\
    .reset_index()
data10
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(data10['Weight (g)'], data10['Avg. Tumor Volume (mm3)'])
ax.set_title('average tumor volume vs. mouse weight for the Capomulin regimen')
ax.set_ylabel('Avg. Tumor Volume (mm3)')
ax.set_xlabel('Mouse Weight')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
import matplotlib.pyplot as plt
from scipy import stats
res = stats.linregress(data10['Weight (g)'], data10['Avg. Tumor Volume (mm3)'])
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(data10['Weight (g)'], data10['Avg. Tumor Volume (mm3)'], 'o', label='original data')
ax.plot(data10['Weight (g)'], res.intercept + res.slope*data10['Weight (g)'], 'r', label='fitted line')
ax.set_title('average tumor volume vs. mouse weight for the Capomulin regimen')
ax.set_ylabel('Avg. Tumor Volume (mm3)')
ax.set_xlabel('Mouse Weight')
ax.legend()
plt.show()
```
